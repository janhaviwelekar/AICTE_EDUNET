```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(df):
    numeric_cols = df.select_dtypes(include=['float64', 'int64'])
    plt.figure(figsize=(8, 6))
    sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
    plt.title("Correlation Heatmap")
    plt.show()
plot_correlation_heatmap(data)
```

```python
def plot_emissions_by_industry(df, top_n=15):
    top_df = df.sort_values(by='Supply Chain Emission Factors with Margins', ascending=False).head(top_n)
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x='Supply Chain Emission Factors with Margins',
        y='2017 NAICS Title',
        data=top_df,
        palette='viridis'
    )
    plt.title('Top Industries by Emission Factors (with Margins)')
    plt.xlabel('Emission Factor (kg CO2e/USD)')
    plt.ylabel('Industry')
    plt.tight_layout()
    plt.show()
plot_emissions_by_industry(data)
```

```python
def plot_margin_analysis(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        x='Supply Chain Emission Factors without Margins',
        y='Supply Chain Emission Factors with Margins',
        hue='Margins of Supply Chain Emission Factors',
        size='Margins of Supply Chain Emission Factors',
        data=df
    )
    plt.title('Emission Factors: With vs Without Margins')
    plt.xlabel('Without Margins')
    plt.ylabel('With Margins')
    plt.tight_layout()
    plt.show()
plot_margin_analysis(data)
```

```python
def plot_boxplot_by_industry(df, top_n=15):
    top_df = df.sort_values(by='Supply Chain Emission Factors with Margins', ascending=False).head(top_n)
    plt.figure(figsize=(12, 6))
    sns.boxplot(
        x='Supply Chain Emission Factors with Margins',
        y='2017 NAICS Title',
        data=top_df,
        palette='Set2'
    )
    plt.title('ðŸ“¦ Box Plot of Emission Factors by Industry')
    plt.xlabel('Emission Factor (kg CO2e/USD)')
    plt.ylabel('Industry')
    plt.tight_layout()
    plt.show()
plot_boxplot_by_industry(data, top_n=15)
```

```python
def plot_violinplot_by_industry(df, top_n=15):
    top_df = df.sort_values(by='Supply Chain Emission Factors with Margins', ascending=False).head(top_n)
    plt.figure(figsize=(12, 6))
    sns.violinplot(
        x='Supply Chain Emission Factors with Margins',
        y='2017 NAICS Title',
        data=top_df,
        palette='coolwarm'
    )
    plt.title('ðŸŽ» Violin Plot of Emission Factors by Industry')
    plt.xlabel('Emission Factor (kg CO2e/USD)')
    plt.ylabel('Industry')
    plt.tight_layout()
    plt.show()
plot_violinplot_by_industry(data, top_n=15)
```
